```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding,Input,RepeatVector
```

```python
train['Script'] = [open(data_folder + os.sep + file, "r").read() for file in train['File_Name']]
test['Script'] = [open(data_folder + os.sep + file, "r").read() for file in test['File_Name']]
```

```python
fn=np.array(train['File_Name'])
x=np.array(train['Script'])
y=np.array(train['Labels'])
```

```python
import string

# cod=toxcom.input_word_ids
for i in range(len(x)):
#     com[i]=com[i].split('n')
    x[i]=' '.join([s.translate(str.maketrans('', '', string.punctuation+str('0123456789'))) for s in x[i].split()])
    x[i]=x[i].split()
# len(x)
# x
```

```python


tokenizer=Tokenizer()
tokenizer.fit_on_texts(x[3])
v=tokenizer.texts_to_sequences(x[3])
if(tokenizer.word_index[x[3][0]]==v[0]):
    print('True')
print(v[0])
print(tokenizer.word_index['slumdog'])
# tokenizer.word_index.items()
```

```python
tok=Tokenizer()
tok.fit_on_texts(x)
seq=tok.texts_to_sequences(x)
```

```python
seqs=[]
for i in range(len(seq)):
  seqs.append(list(set(seq[i])))
```

```python

a=[len(seqs[i]) for i in range(len(seqs))]
max(a)
```

```python
maxlen=max([len(seqen) for seqen in seqs])
maxlen
```

```python
comseq=[]
# for i in range(len(com)):
comseq=pad_sequences(seqs,maxlen,padding='pre')
comseq
```

```python
vocab=len(tok.word_index.items())
vocab
```

```python
c=max(y)
c
```

```python
from sklearn.preprocessing import OneHotEncoder
o=OneHotEncoder()
ys=o.fit_transform(y.reshape(-1,1)).toarray()
```

```python
model=Sequential()
model.add(Embedding(input_dim=vocab+1,output_dim=500,input_length=maxlen,mask_zero=True))
# model.add(LSTM(units=500,return_sequences=True))
model.add(LSTM(units=500))
model.add(Dense(120,activation='sigmoid'))
model.add(Dense(c+1,activation='softmax'))
model.summary()
```

```python

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(comseq,ys)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
cp_callback = [ModelCheckpoint(filepath='/content/drive/My Drive/Colab Notebooks/check-point.ckpt',monitor='loss',
                                                 save_best_only=True,mode='min',
                                                 verbose=1)]
```

```python
from tensorflow.keras import optimizers
rms = optimizers.RMSprop(lr=0.002)
model.compile(optimizer=rms,loss='categorical_crossentropy',metrics=['accuracy'])
```
